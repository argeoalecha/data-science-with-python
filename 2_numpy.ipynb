## NumPy's array class is ``ndarray`` also referred to as ``numpy.array``

### The `ndarray` data structure

The core functionality of **NumPy** is its `ndarray`, for n-dimensional array data structure. It is different from the Python built-in `list` data structure. The **NumPy** arrays are homogenously typed, ie, all elements of an array must be of the same data type.
`ndarray` in **NumPy** is used as the primary container to exchange data between algorithms
>The ``ndarray`` has the following properties
- collection of values
- add, remove, or change elements
- homogenous data type
- multidimensional
- mathematical functions support
- fast and efficient

> The attributes of  ``ndarray``
- .ndim
- .shape
- .size
- .dtype



When it comes to computation, there are three concepts that makes NumPy powerful:

- Vectorization
- Broadcasting
- Slicing and Indexing

### The ``numpy`` n-dimensional array manipulations

````python
# create an array of numbers of range 0 to 35
s_arr = np.arange(36, dtype=None) # dtype in this case can be 'float' if explicitly defined
````

````python
print(s_arr)
print(s_arr.ndim)
print(s_arr.shape) # this is a rank-1 vector or a 1d array; can be a row or a column
````

````python
# transpose a rank-1 array results the same dimension
s_transp = s_arr.T
print(s_transp)
s_transp.shape
````

### ``np.reshape`` and ``np.newaxis``

````python
# reshape a rank-1 array to a n-dim array using reshape
# reshapes the array to a desired matching dimension
s_arr_resh = s_arr.reshape(1, 36) # rows x col or convert to (1,36) for 1 row
# print(s_arr_resh)
print(s_arr_resh.shape)
print(s_arr_resh.T.shape) # transpose
````

````python
# s_row_resh = s_arr.reshape(1,36)
# s_row_resh
# s_row_resh.shape
````

````python
#newsaxis - alternative way of converting rank-1 vector to a ndarray
#a psuedo index which allows the addition of an axis (in row or col); use the slicing operator to recreate the array
s_col_newax = s_arr[:, np.newaxis] 
s_col_newax.shape
````

````python
s_row_newax = s_arr[np.newaxis,:]
s_row_newax.shape
````

## Broadcasting 
- is unique to numpy for array operations

>Broadcasting is a powerful mechanism that allows numpy to work with arrays of different shapes when performing arithmetic operations. Frequently we have a smaller array and a larger array, and we want to use the smaller array multiple times to perform some operation on the larger array.
In most cases, broadcasting is faster, and it is more memory efficient than the equivalent full array operation.


````python
# concat lists in Python
a = [1,2,3]
b = [2,3,4]
print(a + b)
````

````python
#First, to show how we usually do mathematical operations of lists in base Python
a = [1,2,3]
b = [2,3,4]
total = []

for i in range(len(a)):
    total.append(a[i] + b[i])
````

### Images can be represented by ``numpy arrays``

````python
# via matplotlib, a data vis package in Python built over numpy
import matplotlib.pyplot as plt
%matplotlib inline
import matplotlib.image as mpimg
from skimage import data
from scipy.misc import imread, imresize
plt.rcParams['figure.figsize'] = (12, 12)
````

````python
img = mpimg.imread('./images/mayon.jpg')
imgplot = plt.imshow(img)
````

````python
# img
print(img.dtype, img.shape, img.nbytes)
````

````python
img_red = img * [1, 0, 1]
img_yellow = img * [1, 1, 0]
img_blue = img * [0, 1, 1]
````

````python
fig, axs = plt.subplots(1, 3)

axs[0].imshow(img_red)
axs[0].set_axis_off()

axs[1].imshow(img_yellow)
axs[1].set_axis_off()

axs[2].imshow(img_blue)
axs[2].set_axis_off()
````
