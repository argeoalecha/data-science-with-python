#### The normal distribution

<img src="images/normal_dist.png" width="30%" style="margin: 20px;">

````python
x = np.linspace(-100, 100, 300)
mu = 0
sigma = 25

y = (1 / (np.sqrt(2 * np.pi * np.power(sigma, 2)))) * \
    (np.power(np.e, -(np.power((x - mu), 2) / (2 * np.power(sigma, 2)))))
````

````python
plt.grid(b=None, which='major', axis='both', color='grey', linestyle='--', linewidth=0.5)
plt.title('Normal Distibution Curve')
plt.plot(x, y);
````

````python
import scipy.stats as ss

x = np.linspace(-5, 5, 5000)
mu = 0
sigma = 1

y_pdf = ss.norm.pdf(x, mu, sigma) # the normal pdf
y_cdf = ss.norm.cdf(x, mu, sigma) # the normal cdf

plt.plot(x, y_pdf, label='pdf')
plt.plot(x, y_cdf, label='cdf')
plt.legend();
````

## ```seaborn``` package is an enhancement to ``matplotlib``

Seaborn is a Python data visualization library based on matplotlib. It provides a high-level interface for drawing attractive and informative statistical graphics

<img src="images/seaborn.png" width="50%" style="margin: 20px;">

````python
import seaborn as sns
sns.set()

# Load the iris dataset
iris = sns.load_dataset("iris")

# Plot sepal with as a function of sepal_length across days
g = sns.lmplot(x="sepal_length", y="sepal_width", hue="species",
               truncate=True, size=5, data=iris)

# Use more informative axis labels than are provided by default
g.set_axis_labels("Sepal length (mm)", "Sepal width (mm)")
plt.savefig('iris_sns.png', dpi=300)
````
